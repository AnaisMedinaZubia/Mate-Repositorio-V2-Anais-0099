Problema 2: $Y - 5 = 12$
Planteamiento: Tenemos una resta. Para mover el $-5$, aplicamos la operación inversa (suma).Despeje:$Y = 12 + 5$Solución: $Y = 17$



```python
from sympy import symbols, Eq, solve

y = symbols('y')
ecuacion = Eq(y - 5, 12)
solucion = solve(ecuacion, y)
print(f"El valor de Y es: {solucion[0]}")
print("programa realizado por Anais Medina NL = 36")
```

Problema 4: $12 = m + 3m$Planteamiento: Primero debemos agrupar los términos semejantes ($m$ es lo mismo que $1m$).Suma de términos: $12 = 4m$Despeje: El $4$ está multiplicando a $m$, así que pasa dividiendo.$m = 12 / 4$Solución: $m = 3$

```python
from sympy import symbols, Eq, solve

m = symbols('m')
ecuacion = Eq(12, m + 3*m)
solucion = solve(ecuacion, m)
print(f"El valor de m es: {solucion[0]}")
print("programa realizado por Anais Medina NL = 36")
```

Problema 6: $5b + 2b - 3b = 16$Planteamiento: Agrupamos todos los términos con la variable $b$.Operación: $(5 + 2 - 3)b = 16 \rightarrow 4b = 16$Despeje: Pasamos el $4$ dividiendo.$b = 16 / 4$Solución: $b = 4$

```python
from sympy import symbols, Eq, solve

b = symbols('b')
ecuacion = Eq(5*b + 2*b - 3*b, 16)
solucion = solve(ecuacion, b)
print(f"El valor de b es: {solucion[0]}")

print("programa realizado por Anais Medina NL = 36")
```

Problema 8: $7Y - 5 = 3Y + 7$Planteamiento: Llevamos las variables a un lado y los números al otro.Movimiento: $7Y - 3Y = 7 + 5$ (El $3Y$ pasa restando y el $-5$ pasa sumando).Simplificación: $4Y = 12$Despeje: $Y = 12 / 4$Solución: $Y = 3$

```python
from sympy import symbols, Eq, solve

y = symbols('y')
ecuacion = Eq(7*y - 5, 3*y + 7)
solucion = solve(ecuacion, y)
print(f"El valor de Y es: {solucion[0]}")


print("programa realizado por Anais Medina NL = 36")
```

Problema 10: $8n - 6 = 4n + 10$Planteamiento: Agrupamos términos semejantes.Movimiento: $8n - 4n = 10 + 6$Simplificación: $4n = 16$Despeje: $n = 16 / 4$Solución: $n = 4$

```python
from sympy import symbols, Eq, solve

n = symbols('n')
ecuacion = Eq(8*n - 6, 4*n + 10)
solucion = solve(ecuacion, n)
print(f"El valor de n es: {solucion[0]}")

print("programa realizado por Anais Medina NL = 36")
```

Problema 12: $10b + 8 = 4b + 10$Planteamiento: Agrupamos variables y constantes.Movimiento: $10b - 4b = 10 - 8$Simplificación: $6b = 2$Despeje: $b = 2 / 6$Solución: $b = 1/3$ o $0.33...$

```python
from sympy import symbols, Eq, solve

b = symbols('b')
ecuacion = Eq(10*b + 8, 4*b + 10)
solucion = solve(ecuacion, b)
# Esto devolverá la fracción 1/3
print(f"El valor de b es: {solucion[0]}")

print("programa realizado por Anais Medina NL = 36")
```

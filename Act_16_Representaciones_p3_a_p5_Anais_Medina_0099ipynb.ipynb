#### Problema 3
 Análisis y Resolución MatemáticaProblema 3: Un número al que se le resta sietePlanteamiento: Sea $x$ el número desconocido. La frase "se le resta siete y da 18" se traduce como:$$x - 7 = 18$$Despeje:Sumamos 7 en ambos lados para dejar a la $x$ sola.$x = 18 + 7$Resultado: $x = 25$

```python
# Problema 3: x - 7 = 18
resultado = 18
resta = 7

# Despeje: x = resultado + resta
numero = resultado + resta

print(f"El número buscado en el Problema 3 es: {numero}")
```

#### problema 5
Datos:Largo ($L$): mide 5 metros más que el ancho ($L = A + 5$).Ancho ($A$): es nuestra incógnita base.Perímetro total ($P$): 50 metros.Planteamiento de la ecuación:El perímetro de un terreno rectangular es la suma de sus cuatro lados (2 veces el largo más 2 veces el ancho):$$P = 2L + 2A$$Sustituimos el largo ($L$) por su equivalencia ($A + 5$):$$50 = 2(A + 5) + 2A$
 $Despeje de la variable:Distribuimos el 2: $50 = 2A + 10 + 2A$
$Sumamos las $A$: $50 = 4A + 10
$Restamos 10 en ambos lados: $50 - 10 = 4A \rightarrow 40 = 4A$Dividimos entre 4: $A = 40 / 4$Resultado Ancho ($A$): 10 metros.Resultado Largo ($L$): $10 + 5 = 15$ metros.

```python
# Problema 5: Terreno con perímetro de 50m
# Ecuación: 50 = 2*(A + 5) + 2*A -> 50 = 4A + 10

perimetro_total = 50
diferencia_largo = 5

# Paso 1: Restar el exceso del largo multiplicado por 2 (los +10 de la ecuación)
# Paso 2: Dividir entre 4 (el número de veces que aparece 'A' en el perímetro)
ancho = (perimetro_total - (2 * diferencia_largo)) / 4

# El largo es el ancho más la diferencia
largo = ancho + diferencia_largo

print(f"Resultados del terreno:")
print(f"El ancho mide: {ancho} metros")
print(f"El largo mide: {largo} metros")

# Comprobación (opcional)
print(f"Comprobación del perímetro: (2*{largo}) + (2*{ancho}) = {2*largo + 2*ancho}m")
print("Prorama realizado por Anais medina NL = 36")
```
